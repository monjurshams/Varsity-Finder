```python
import webbrowser
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time 
```

## First we need to declare a header variable containing user agent for proxy, this is required for the websites that have firewall. These universities are among those that have the acceptance rate higher than 50%.

```python
headers = {'User-Agent':'Mozilla/5.0'}
browser = webdriver.Chrome()
browser.get('https://www.usnews.com/best-colleges/rankings/engineering-doctorate?_sort=acceptanceRate&_sortDirection=desc&acceptanceRateMin=50')
browser.maximize_window()
```

## Now we find the university names from the site using the css selector grabbed from the site. As the site needed to be scrolled down to fetch every name, we used selenium module to scrape data. Selenium gives flexibility to control the browser from outside and it works better with javascript based website.

```python
name_list = []

while True:
    
    load_button = browser.find_element_by_css_selector("#search-content > div > div:nth-child(1) > div.Cell-sc-1abjmm4-0.ezHWed > div.pt1 > div > div.LoadMoreWrapper__Container-zwyk5c-0.cMoThM > div > button > div")
    uni_name = browser.find_elements_by_css_selector("h3")
    
    for item in uni_name:
        if item.text in name_list:
            
            continue
            
        else:    
            name_list.append(item.text)
            
            
    browser.execute_script("arguments[0].scrollIntoView();",load_button)
    time.sleep(4)
    
```

## Then we exported the dataset to a csv file

```python
import pandas as pd
df = pd.DataFrame(name_list)
df
df.to_csv("Output.csv")
```

## Using the same process, we fetched the university list that accept gre score range 301-310. But this time we used Beautiful Soup module instead of selenium. Because it was easier to implement as the website didnt need that much controlling from outside.

```python
dict_list=[]
base_url = "https://studyabroad.shiksha.com/ms-colleges-for-gre-score-301-to-310-"
for page in range(0,5):
     
    req = requests.get(base_url+str(page))
    cont = req.content
    soup = BeautifulSoup(cont,"html.parser")
    #mydivs = soup.findAll("#categoryPageListing_tupleId_204150 > div.tuple-detail > div.tuple-title > p:nth-child(2) > a")
    #print(mydivs)
       # print(names.text)
    for hit in soup.findAll(attrs={'class' : 'tuple-title'}):
        if hit.text not in dict_list:
            dict_list.append(hit.text)
        print (hit.text) 
    
```

## Then we cleaned the data by removing useless information

```python

for item in dict_list:
    if item.startswith("\nMS ") or item.startswith('\nMA ') or item.startswith('\nM.S. ') or item.startswith('\nM.S.in ') or item.startswith('\nM.S ') or item.startswith('\nMaster of ') or item.startswith('\nCivil ') or item.startswith('\nMechanical ') or item.startswith('\nManagement ') :
        dict_list.remove(item)

   
print(dict_list)
```

```python
uni_gre=[]
for item in dict_list:
    new_names = item[10:]
    uni_gre.append(new_names)
```

```python
gre_list = []
for item in uni_gre:
    gre_list.append(item.strip())
gre_list
```

## Finally I exported the list to a csv file

```python
df2 = pd.DataFrame(gre_list)
df2.to_csv("GRE(300-310).csv")
```
